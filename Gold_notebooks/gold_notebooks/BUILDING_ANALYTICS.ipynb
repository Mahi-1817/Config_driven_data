```python
%sql
CREATE OR REPLACE VIEW config_catalog.gold.sales_summary AS
SELECT
    c.customer_name,
    p.product_name,
    SUM(f.amount) AS total_sales
FROM config_catalog.gold.fact_sales f
JOIN config_catalog.gold.dim_customers c
    ON f.customer_sk = c.customer_sk
JOIN config_catalog.gold.dim_products p
    ON f.product_sk = p.product_sk
GROUP BY c.customer_name, p.product_name;
```

```python
%sql
CREATE OR REPLACE VIEW config_catalog.gold.monthly_sales AS
SELECT
    DATE_TRUNC('month', order_date) AS sales_month,
    SUM(amount) AS revenue
FROM config_catalog.gold.fact_sales
GROUP BY DATE_TRUNC('month', order_date);
```

```python
%sql
OPTIMIZE config_catalog.gold.fact_sales;
OPTIMIZE config_catalog.gold.dim_customers;
```

```python
%sql
OPTIMIZE config_catalog.gold.fact_sales
ZORDER BY (customer_sk, product_sk);
```
